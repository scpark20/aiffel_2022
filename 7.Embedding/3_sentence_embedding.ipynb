```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = ['Many students are studying at the high school.',
             'Many teachers are working at the high school.',
             'Many students are not studying at the high school.',      
             'You can find plagiarized documents through sentence embedding',
             'The sentence embedding model was built with Transformer.']

sentence_embeddings = model.encode(sentences)
```

```python
corr = sentence_embeddings @ sentence_embeddings.T

import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(corr, annot = True)
plt.show()
```
